```python
import pandas as pd
import numpy as np
import os

#hide warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#load trade list
df = pd.read_csv('trade_list.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.dropna(inplace=True)
df.set_index('Date',inplace=True)
df = df.groupby('Stock')


```

```python
def trades(buys,sells):
    buys_and_sells = pd.DataFrame()
    buys_and_sells['buy_date'] = buys.index
    buys_and_sells['sell_date'] = sells.index
    buys_and_sells['buy_price'] = buys['Price'].values
    buys_and_sells['sell_price'] = sells['Price'].values
    buys_and_sells['profit'] = (buys_and_sells['sell_price']-buys_and_sells['buy_price'])/buys_and_sells['buy_price']*100
#     print(buys_and_sells)

    buys_and_sells.set_index('sell_date',inplace=True)
    buys_and_sells.drop('buy_price',1,inplace=True)
    buys_and_sells.drop('sell_price',1,inplace=True)
    
    print("\n")
    
    #print montly and yearly profits
    print("~MONTHLY~")
    print(buys_and_sells.resample('M').sum())
    
    print("\n")
    
    print("~YEARLY~")
    print(buys_and_sells.resample('Y').sum())
    
    print("#######################################################")
    print("\n")
```

```python
stock_list = [s for s,d in df ]
print(stock_list)
```

```python
for s in stock_list:
    print(s)
    data = df.get_group(s)
    data.sort_values('Date',inplace=True)

    #separate buys and sells
    buys = data[data['Action']==1]
    sells = data[data['Action']==2]
    print(f"buys:{len(buys)} sells:{len(sells)}")
    
    #align length of buys and sells
    buys = buys[:len(sells)]
    print(f"trades:{len(buys)}")
    
    trades(buys,sells)

# print(buys,sells)
```
